```python
import numpy as np
RI_dict = {1: 0, 2: 0, 3: 0.58, 4: 0.90, 5: 1.12, 6: 1.24, 7: 1.32, 8: 1.41, 9: 1.45}
def AHP(A_arr):     #求特征向量
    A = np.array(A_arr)
#将矩阵列向量进行单位化0
    a_sum0 = A.sum(axis=0)
    B = A / a_sum0  
    

    b_sum = B.sum(axis=1)
    W = b_sum.sum()
    w_arr = []
    for w in b_sum:
        w_arr.append(w/W)
    print('特征向量为: %s' % w_arr)
    #一致性检验
    AW = []
    for a in A :
        aa = a * w_arr
        AW.append(aa.sum())
    result = np.array(AW) / np.array(w_arr)
    row = result.shape[0]
    Max = result.sum()/row
    print('最大特征值λMax: %s' % Max)
    CI = (Max - row) / (row - 1)
   
    CR = CI / RI_dict[row]
    print('CR: %s' % CR)
    if (CR<0.1): print('通过一致性检验！')
    else: print('未通过一致性检验！')
```
